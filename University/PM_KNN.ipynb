```python
import numpy as np
from pylab import rcParams
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets
import pandas as pd
```

```python
rcParams['figure.figsize'] = 12, 8
```

```python
muB = (1,0)
muO = (0,1)
n_of_dots = 25
sc=0.5
h=0.01 # for grid
y_min=-3
y_max=3
x_min=-3
x_max=3
cmap = ListedColormap(['#ffe4b2','#b2b2ff' ])
```

```python
np.random.seed(7)
```

```python
MeansB = list(zip(np.random.normal(loc=muB[0], scale=0.5, size=10), np.random.normal(loc=muB[1], scale=0.5, size=10)))
MeansO = list(zip(np.random.normal(loc=muO[0], scale=0.5, size=10), np.random.normal(loc=muO[1], scale=0.5, size=10)))
```

```python
Blues = []
Oranges = []
num_of_samples = 10
for i in range(10):
    Blues.extend(list(zip(np.random.normal(loc=MeansB[i][0], scale=sc, size=n_of_dots), np.random.normal(loc=MeansB[i][1], scale=sc, size=n_of_dots), np.ones(n_of_dots))))
    Oranges.extend(list(zip(np.random.normal(loc=MeansO[i][0], scale=sc, size=n_of_dots), np.random.normal(loc=MeansO[i][1], scale=sc, size=n_of_dots), np.zeros(n_of_dots))))
```

```python
df = pd.DataFrame(columns=['X1','X2','Y'], data=Blues+Oranges)
```

```python
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
```

```python
n_neighbors = 2
clf = neighbors.KNeighborsClassifier(n_neighbors, weights='distance')
clf.fit(df[['X1','X2']], df.Y)
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx, yy, Z, cmap=cmap)
plt.scatter(df[df.Y==1].X1, df[df.Y==1].X2, color='blue')
plt.scatter(df[df.Y==0].X1, df[df.Y==0].X2, color='orange')
```
